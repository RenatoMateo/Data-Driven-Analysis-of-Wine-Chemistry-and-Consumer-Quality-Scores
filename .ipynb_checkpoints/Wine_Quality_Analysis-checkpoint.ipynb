### Problem Statement for Wine Quality Analysis

The wine market is a major segment of the global alcohol industry, with a market value of USD 339.53 billion as of now and a forecast to reach 456.76 billion in 2028, according to Fortune Business Insights. Traditionally, wine quality has been evaluated by human experts via sensory tests, but this method is subjective, inconsistent, and influenced by factors that cannot be directly measured, even in standard chemical analyses. The goal of this study is to determine whether wine quality scores can be reliably segmented, profile and predicted using only physicochemical properties available in laboratory measurements.

Important questions to consider in the Analysis Section of this project:

- What physicochemical characteristics and chemical profiles correlate with higher wine quality scores? Also, what are the specific features that differentiate high-quality wines from lower-quality ones?

- Which physicochemical parameters have the greatest impact on wine quality and what are the threshold levels that should be controlled and monitored in production and quality control?

- How accurately can machine learning models predict wine quality based on physicochemical data, and what insights do these models offer about the connection between chemistry and perceived quality?

Datasets Source: Cortez, P., Cerdeira, A., Almeida, F., Matos, T., & Reis, J. (2009). Wine Quality [Dataset]. UCI Machine Learning Repository. https://doi.org/10.24432/C56S3T.

***Chemical Insight: Correlation Package***

***1. Alcohol:***
Correlations:
- Strong negative with density (-0.5)
- Slight negative with volatile acidity(-0.20), chlorides (-0.22), and total sulfur dioxide (-0.21)

***Why:***
Ethanol is less dense than water, so as alcohol increases, density drops - chemistry Reaction by default. 
Higher‑alcohol wines also tend to come from riper grapes and cleaner fermentations, which naturally produce:
- lower volatile acidity
- lower chlorides
- lower need for SO₂ additions

So alcohol acts as a proxy for ***ripeness and fermentation quality***.
Meaning is an indicator of ***how efficient*** was the timing of the ***harvesting*** and ***fruit process to storage***. 


***2. Sulphates***

Correlations:
- Slight positive with chlorides (0.37).
- Slight positive with density (0.15).
- Slight negative with volatile acidity (-0.26).

Why:

- Sulphates are added as stabilizers and antioxidants. Also, grapes comes with natural levels of sulphates.
- Their negative correlation with volatile acidity suggests that wines with more sulphates tend to have better microbial control, reducing acetic acid formation.
- The positive link with chlorides and density reflects that both are dissolved inorganic ions, contributing modestly to total dissolved solids.One come with the other since both comes from minerals ions in the soild. 

Added Sulphates are part of a normal winemaking process to keep the chemical reactions and quality product stable.




***3. Volatile Acidity.***

Correlations:
- Slight negative with alcohol (-0.20).
- Slight negative with sulphates (-0.26).

Why:

- Volatile acidity increases when fermentation is stressed or poorly controlled.
- Higher alcohol and higher sulphates both indicate cleaner, healthier fermentations, so volatile acidity tends to be lower in those wines.

Volatile acidity becomes a benchmark of fermentation quality. 


***4. Total Sulfur Dioxide.***

Correlations:
- Slight negative with alcohol (-0.21).
- Slight positive with density (0.07).
- Weak relationships with the other variables

Why:

- Wines with higher alcohol often require less SO₂ for microbial stability, explaining the negative correlation.
- SO₂ contributes to dissolved solids, so it nudges density upward.
- SO₂ is a winemaking decision, not a natural grape component. So, it doesn’t follow natural chemical patterns.

Its weak correlations with other variables reflect that ***SO₂ additions*** depend heavily on ***winemaking decisions***, not just grape chemistry.

### Analysis Section.

### Segmentation

After looking at how chemical composition and consumer scores are related, as well as the correlations between the physicochemical variables themselves, we first moved on to address the initial research question of the project:

- What physicochemical characteristics and chemical profiles correlate with higher wine quality scores?Also, what are the specific features that differentiate high-quality wines from lower-quality ones?

To address this issue, Principal Component Analysis (PCA) was employed to simplify the data structure and mitigate the excessive multicollinearity detected among several variables. After that, cluster analysis was performed through the K-Nearest Neighbors (KNN) technique to categorize wines with similar chemical patterns.

This further helped in locating clusters that were more associated with higher quality scores and also allowed the chemical compositions of these clusters to be compared with those of the lower-scoring groups.

***PCA on Red Wine***

```python
#PCA Scores for Red Wine df
# Select Features for PCA
features = df_red_wine.columns[:-1] # all except last column
x = df_red_wine[features].values

# Standardize the features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Apply PCA
pca = PCA (n_components=2)
x_red_pca = pca.fit_transform(x_scaled)

# Get PCA components (loadings)
loadings = pca.components_.T  # transpose to match features x components

# Create DataFrame
pca_loadings = pd.DataFrame(loadings, index=features, columns=['PC1', 'PC2'])

# Round values for readability
pca_loadings = pca_loadings.round(6)

print(pca_loadings)
```

***Insight Red Whine:***

PC1 Grape Chemistry Profile: 

Strong Positive Loading:
- fixed acidity (+0.49)
- citric acid (+0.46)
- density (+0.39)
- sulphates (+0.24)
- chlorides (+0.21)

Strong Negative Loading: 
- pH (–0.44)
- volatile acidity (moderate negative, –0.23)

In the red wine dataset, PC1 captures the natural chemistry of the grape, separating wines that are more acidic, denser, and richer in minerals (high fixed acidity, citric acid, density, chlorides, sulphates, and low pH) from wines that are riper, softer, and higher in alcohol. This PC1 tell the story o chemical composition of grapes. High PC1 wines have less ripe grapes and low PC1 have more ripe grapes. 

----
PC2 Processing & Preservation profile:

Strong Positive Loading:
- Overall Total and Free SO2 (0.57)
- volatile acidity (0.27)
- residual sugar (0.27)

Strong Negative Loading: 
- Alcohol (–0.39)

PC2 reflects the degree of preservation, stabilization, and fermentation control applied to the wine. High PC2 wines require more SO₂, more correction, more protection, slightly more residual sugar → wines that needed more stabilization. Low PC2 Cleaner fermentations, lower SO₂, higher alcohol lead to wines that required less intervention.

***PCA on White Wine***

```python
#PCA Scores for White Wine df
# Select Features for PCA
features = df_white_wine.columns[:-1] # all except last column
x = df_white_wine[features].values

# Standardize the features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Apply PCA
pca = PCA (n_components=2)
x_white_pca = pca.fit_transform(x_scaled)

# Get PCA components (loadings)
loadings = pca.components_.T  # transpose to match features x components

# Create DataFrame
pca_loadings = pd.DataFrame(loadings, index=features, columns=['PC1', 'PC2'])

# Round values for readability
pca_loadings = pca_loadings.round(6)

print(pca_loadings)
```

***Insight White Wine:***

PC1 Grapes ripe quality: 

Strong Positive Loading:

PC1 reflects how ripe the grapes were and how much stabilization the wine required. High‑PC1 wines come from less ripe grapes, with lower alcohol and higher density, requiring more SO₂ and more correction. Low‑PC1 wines come from riper grapes, ferment cleanly, produce more alcohol, and need less intervention.

***Cluster Analysis for red and white wine***

```python
# Selecting number of ideal clusters for segmentation for red wine df. 
# Red Wine Dataset: 
wcss = []
# Use only numeric columns
x = x_red_pca

#Statistical significance of the k selection. 
silhouette_scores = []

for i in range(2, 11):  # silhouette needs at least 2 clusters
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)
    labels = kmeans.fit_predict(x)
    score = silhouette_score(x, labels)   # <-- use silhouette_score, not silhouette_scores
    silhouette_scores.append(score)
    print(f"Clusters: {i}, Silhouette Score: {score:.3f}")

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title("The Elbow Method- Red Wine Cluster")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
    
```

```python
# White Wine Dataset: 
wcss = []
x = x_white_pca

#Statistical significance of the k selection. 
silhouette_scores = []

for i in range(2, 11):  # silhouette needs at least 2 clusters
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)
    labels = kmeans.fit_predict(x)
    score = silhouette_score(x, labels)   # <-- use silhouette_score, not silhouette_scores
    silhouette_scores.append(score)
    print(f"Clusters: {i}, Silhouette Score: {score:.3f}")

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title("The Elbow Method- White Wine Cluster")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
    
```

***Insight on number of Cluster Selection:*** 
- Red Wine and White wine: Based on plot of Elbow method on cumulative WCSS we infer that 3 is the ideal number for both datasets. This is also supported by the silhouette_scores that indicates 3 with significant number of 0.377 and 0.378 for red and white wine respectevely. 

***k-Means Model for Clustering on Red and white wine df.***



```python
# Choose the best k from elbow + silhouette RED WINE. 
best_k = 3
kmeans_final = KMeans(n_clusters=best_k, init="k-means++", random_state=42)
cluster_labels = kmeans_final.fit_predict(x_red_pca)
df_red_wine["Clusters"] = cluster_labels
# Add cluster labels to PCA dataframe
red_pca_df = pd.DataFrame(x_red_pca, columns=["PC1", "PC2"])
red_pca_df["Cluster"] = cluster_labels

print(red_pca_df.head())

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=red_pca_df,
    x="PC1",
    y="PC2",
    hue="Cluster",
    palette="Set2",
    s=80
)

plt.title("Red Wine Clusters on PCA Space")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.savefig('Cluster_Red.png') 
plt.show()
```

```python
# Choose the best k from elbow + silhouette WHITE WINE
best_k = 3
kmeans_final = KMeans(n_clusters=best_k, init="k-means++", random_state=42)
cluster_labels = kmeans_final.fit_predict(x_white_pca)
df_white_wine["Clusters"] = cluster_labels

# Add cluster labels to PCA dataframe
white_pca_df = pd.DataFrame(x_white_pca, columns=["PC1", "PC2"])
white_pca_df["Cluster"] = cluster_labels
white_pca_df.head()

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=white_pca_df,
    x="PC1",
    y="PC2",
    hue="Cluster",
    palette="Set2",
    s=80
)

plt.title("White Wine Clusters on PCA Space")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.savefig('Cluster_White.png') 
plt.show()
```

***Selection of Threshold Variables for Red and White Wines***

To set proper quality thresholds for red and white wines, we first pinpointed a range of key physicochemical variables for each type of wine. These variables were chosen because they represent very highly correlated features with consumer score and are key drivers of PCA. In addition, our decision to organize these variables was motivated by the fact that they collectively represent three dimensions of wine quality:

***Positive markers of grape ripeness:***

***1. Alcohol (% v/v)***
- The strongest positive correlation with consumer score.
- Reflects grape ripeness and complete fermentation.
- Higher alcohol is consistently associated with higher liking.

***2. Sulphates (g/L)***
- Significant strongest positive correlation with consumer score.
- Acts as a stabilizing agent and contributes to structure and mouthfeel.
- Grapes comes with natural levels of sulphates.

***3. Citric Acid (g/L)***
- Third strongest positive correlation with consumer score.
- Enhances freshness and contributes to balance.
- Grapes Natural level of Acid and indicator of ripeness.
  

***Elements related to fermentation, maceration, and storage processes:***

***4. Density (g/cm³)***
- A major positive driver of PC on both type of wines.
- Summarizes fermentation completeness, residual sugar, and stabilization effort.
- Lower density indicates cleaner fermentation and better process control.

***5. Total Sulfur Dioxide (mg/L)***
- A key process variable, contributing to PC drivers.
- Higher levels indicate greater stabilization effort.
- Controlled SO₂ is essential for quality.

***6. Residual Sugar (g/L)***
- Another PC1 driver, linked to Ripe Fruit.
- Lower RS indicates dry, stable wines with fewer interventions.
- Better fermentation process lower residual sugar turn into alcohol. 


***Negative markers of underripeness or chemical defects:***

***7. Volatile Acidity (g/L)***
- significant negative correlation with consumer score.
- High VA indicates microbial stress and leads to vinegar‑like or solvent aromas.
- Lower VA is essential for clean, fault‑free wines.It indicates fault on grape or maseration process. 

***8. Chlorides (g/L)***
- Second strongest negative correlation with consumer score.
- Also a positive driver of PC, linked to less ripe or more stressed fruit.
- Unripe grapes have naturally more chlorides. 

***9. Ph.***
- Second strongest positive correlation with consumer score.
- Overall indicator of Acidic on grape ripenss.
- Low Ph indicates high leves of acidity in wine control.

With these variables, we can assess the extent to which each of the different facets of grape maturity and winemaking practices influences the quality of the wine and therefore consumer scores.
